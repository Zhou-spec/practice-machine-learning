```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from ISLP import load_data
```

```python
boston = load_data('Boston')
boston.head(10)
```

```python
import statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor

variance_inflation_factor(boston, 1)

for i in range(boston.shape[1]):
    print(f"this is the {i}-th VIF")
    print(variance_inflation_factor(boston, i))
```

```python
X = boston['lstat']
intercept = pd.Series(np.ones(boston.shape[0]))
X = pd.concat([X, intercept], axis = 1)
X.columns = ['lstat', 'intercept']
y = boston['medv']
y.head(10)
```

```python
import statsmodels.api as sm 

model = sm.OLS(y, X)
result = model.fit()
result.params
```
